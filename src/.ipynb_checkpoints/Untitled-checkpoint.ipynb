```python
import datetime as dt
start_date = dt.datetime(2020, 7, 1, 8, 45, 0)
end_date = dt.datetime(2023, 8, 1, 13, 45, 0)
trade_start = '08:45'
trade_end = '13:45'
dates = pd.date_range(start=start_date, end = end_date, freq = 'T')
df = pd.DataFrame(index=dates[(dates.time >= pd.to_datetime(trade_start).time()) &
                              (dates.time <= pd.to_datetime(trade_end).time())])
# df['a'] = [1]*len(df.index)
# df['b'] = [1]*len(df.index)
# df['ts'] = df.index
df['High'] = [-1]*len(df.index)
df['Low'] = [-1]*len(df.index)
df['Open'] = [-1]*len(df.index)
df['Amount'] = [-1]*len(df.index)
df['Volume'] = [-1]*len(df.index)
df['Close'] = [-1]*len(df.index)
df.to_csv("test.csv", index = False)
df
```

```python
data_path = "../data/future/"
for file in os.listdir(data_path):
    df_all = df.copy()
    stock_id = file.split(".")[0]
    print("Load ", stock_id)
    df_single = pd.read_csv(os.path.join(data_path, file))
    df_single["ts"] =  pd.to_datetime(df_single["ts"])
    df_single.set_index("ts", inplace = True)
    df_single.index.name = None
    df_single = df_single[~df_single.index.duplicated(keep='first')]
#     print(df_single[df_single.index.duplicated()])
    df_all.update(df_single)
    df_all['ts'] = df_all.index
    first_column = df_all.pop('ts')
    df_all.insert(0, 'ts', first_column)
    df_all.to_csv("../data/future_test2/"+stock_id+".csv", index = False)
    
```
